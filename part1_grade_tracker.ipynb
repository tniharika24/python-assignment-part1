```python
#Task 1 — Data Parsing & Profile Cleaning

# given raw data
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

students = []

for s in raw_students:

    #cleaning name
    name = s["name"].strip()
    name = name.title()

    #converting roll
    roll = int(s["roll"])

    #converting marks
    marks = []
    parts = s["marks_str"].split(",")

    for p in parts:
        marks.append(int(p.strip()))

    #checking valid name
    valid = True
    for word in name.split():
        if word.isalpha() == False:
            valid = False

    if valid:
        print(f"{name} ✓ Valid name")
    else:
        print(f"{name} ✗ Invalid name")

    #storing cleaned data
    students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

    #f-string formatted profile card
    print("=" * 32)
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("=" * 32)


#printing for roll 103
for s in students:
    if s["roll"] == 103:
        print(s["name"].upper())
        print(s["name"].lower())
```

```python
#Task 3 — Class Performance Summary

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

pass_count = 0
fail_count = 0

top_name = ""
top_avg = 0

total_avg = 0

print("Name              | Average | Status")
print("-------------------------------------")

for data in class_data:
    name = data[0]
    marks = data[1]
    avg = round(sum(marks) / len(marks), 2)
    total_avg += avg

    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    if avg > top_avg:
        top_avg = avg
        top_name = name

    print(f"{name:<18} | {avg:^7} | {status}")


print("\nPassed:", pass_count)
print("Failed:", fail_count)

print("Topper:", top_name, top_avg)

class_avg = round(total_avg / len(class_data), 2)
print("Class Average:", class_avg)
```
